```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df_train = pd.read_csv(filepath_or_buffer=r"titanic/train.csv")
df_train
```

```python
y = df_train.Survived
df_train = df_train.drop("Survived",1)

df_train.Sex.replace(['male', 'female'], [1, 0],inplace=True)
df_train.Embarked.replace(['S','C','Q'],[1,2,3],inplace=True)

df_train
```

```python
df_train = df_train.drop("Cabin",1)
```

```python
df_train = df_train.drop("Name",1)
```

```python
df_train = df_train.drop("Ticket",1)
```

```python
df_train = df_train.dropna(axis=0)
```

```python
df_train
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(df_train,y)
```

```python
X_test = pd.read_csv(filepath_or_buffer=r"titanic/test.csv")
```

```python
X_test = X_test.drop("Cabin",1)
```

```python
X_test = X_test.drop("Name",1)
X_test.Sex.replace(['male', 'female'], [1, 0],inplace=True)
X_test.Embarked.replace(['S','C','Q'],[1,2,3],inplace=True)
```

```python
X_test = X_test.drop("Ticket",1)
```

```python
X_test[X_test.isna()] = X_test.Age.mean()
```

```python
y_pred = model.predict(X_test)
```

```python
output = pd.DataFrame(X_test.PassengerId)
```

```python
output["Survived"] = y_pred
```

```python
output = output.set_index("PassengerId")
```

```python
output.to_csv(r"titanic/output.csv")
```
